# Import Relevant Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Raw Data and Find Vinyl Peak

```python
# Import t0 raw data
df = pd.read_csv('t60_1.txt', delimiter = '\t') # Tab delimited.
df = df.drop(df.columns[0:3], axis=1) # Drop 1st 3 columns.
df
```

```python
# Find out Vinyl Peak Wavenumber = ~1630 cm-1 Region

df.iloc[:, 375:385].plot()
plt.show() 
# Plot all columns in the region to see which intensity is the highest. Pick that index location.
# Wavenumber at 1628.104424 cm-1 is the vinyl peak with the highest intensity.
```

```python
# df.loc[:,'1629.209334'] --> When using loc.
df.columns.get_loc('1629.245625') #--> Use the get_loc command on the column index to find the index location. 
# df.iloc[:,378] --> When using iloc

# Column index 378 is equivalent to column label 1628.659771. Use relevant parameter for iloc and loc respectively.
```

# Filter Relevant Spectra

```python
# Zoom into Relevant Region
df.loc[50:150,'1629.245625'].plot()
plt.xlabel('Spectral Index')
plt.ylabel('Intensity')
plt.title('0 Hours')
plt.show()
```

```python
# Extract Data From Steady State Region
# Zoom into Relevant Region
df.loc[100:110,'1629.245625'].plot()
plt.xlabel('Spectral Index')
plt.ylabel('Intensity')
plt.title('0 Hours')
plt.show()
```
