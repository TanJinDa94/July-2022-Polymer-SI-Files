```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import os
```

```python
'''
Read through 27 excel files using glob module. Make sure the files are ordered according to condition and residence time.
'''


file_list = glob.glob(r'*.xlsx')
# Go to directory and return a list of all the file names in that location.

file_list = sorted(file_list, key=os.path.getmtime)
# Sort files according to the time that it was created.

indices = [x for x in range(len(file_list))]
d = dict(zip(indices, file_list))
d
# Inspect the order of the file_list. Indices coupled with file name.
```

```python
df = pd.read_excel(file_list[0]) # Read 1st file

df # DataFrame needs cleaning to extract MW and Mole Fraction data
```

```python
x = list(df.iloc[:,11].dropna()) # Drop all empty values
x = x[1:] # Drop 1st element 'Molecular Weight' of list

M_i = [] # Create a list of Log M values
for i in x:
    M_i.append(i)


n_i = list(df.iloc[:,14].dropna()) # Drop all empty values
n_i = n_i[1:] # Drop 1st element 'Normalized Mole Fraction' of list. n_i list represents the Normalized Mole Fraction.

'''
Create the scaffold of a DataFrame so that each file can be iteratively read and then appended in rows.
The DataFrame will contain the log molecular weights as its header, and the values of n_i in each file as its values.
The length of each row is fixed to 1499 values, owing to the boundary set during data extraction using OmniSEC.
Each row will represent an MWD when plotted against the header.
'''

df_ = dict(zip(M_i,n_i)) # zip the M_i and n_i lists together and create a dictionary
df_ = pd.DataFrame(df_, index=[0]) # Create a Dataframe from the dictionary and set index to log MW.
df_ # Inspect df. Header must be the absolute molecular weight, each row represents the n_i value of each file at that MW.
```

```python
for i in range(len(file_list)): # i ranges from 0 to 26 (27 MWD files)
    df1 = pd.read_excel(file_list[i]) # Dummy df1 object stores the unedited excel file DataFrame within the loop
    n_i = list(df1.iloc[:,14].dropna()) # Drop missing values and create a list
    n_i = n_i[1:] # Drop 'Normalized Mole Fraction'
    df_.loc[i] = n_i # Overwrite the ith row of df with n_i

df_ = df_.div(df_.sum(axis=1), axis=0) # Normalize MWD to sum to 1
df_
```

```python
'''
Re-index the rows of df into the corresponding condition and residence time for book-keeping purposes.
'''

index =[]
for i in file_list:
    i = i.replace('.xlsx','') # Replace substring '.xlsx' with ''
    index.append(i) 

df_.index = index
df_
```

```python
i=0

for index, item in df_.iterrows():
    i+=1
    x = df_.columns
    y = item
    plt.figure(figsize = (15,270))
    plt.subplot(27,1,i)
    plt.plot(x,y)
    plt.title(str(index), fontsize=16)
    plt.xlabel('Log Molecular Weight',size=16)
    plt.ylabel('Normalized Weight Fraction ($w_{i}$)',size=16)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
plt.tight_layout()
plt.show()
```

```python
import numpy as np

l = np.arange(0,15,3) # Create a list which encodes the index location for every 3 rows.
l

for k,i in enumerate(l):
    
    df_dummy = df_.iloc[i:i+3,:]
    x = df_.T.index
    y = df_dummy.mean()
    stderr = df_dummy.std()
    
    plt.figure(figsize=(15,6))
    plt.title('C15 T30 Organic Segment ' + str(k+1), fontsize = 16)
    plt.plot(x, y, alpha = 0.5, label='Mean')
    plt.fill_between(x, (y-stderr), (y+stderr), color='b', alpha=0.1, label='Spread')
    plt.legend(fontsize= 16)
    plt.xlabel('Log Molecular Weight',size=16)
    plt.ylabel('Normalized Weight Fraction ($w_{i}$)',size=16)
    plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0), useMathText=True)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    
    plt.show()
```
